Groupby


```python
import pandas as pd

data = {'Department': ['Sales', 'Sales', 'HR', 'HR', 'IT', 'IT', 'IT'],
        'Employee': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace'],
        'Salary': [50000, 60000, 55000, 52000, 70000, 75000, 72000]}

df = pd.DataFrame(data)
print(df)
```

```python
df.groupby('Department')['Salary'].sum()
```

```python
var = flower_details.groupby('species').count()
```

```python
print(var)
```

```python
var_1 = flower_details.groupby('sepal_length').count()

print(var_1)
```

```python
print(flower_details.groupby('species').sum())
```

```python
print(flower_details.groupby('species').mean())
```

```python
print(flower_details.groupby('sepal_length').sum())
```
